```julia
rc("text", usetex=true)
rc("text.latex", preamble= "\\usepackage{amsmath}")
rc("font", family="serif")
pygui(false)

k = 2 * π
arr_length = 100
x = collect(range(-0.25, 1.25, length=arr_length))
# dens_1 = @. sin(k*x)^2
dens_1 = @. exp(-80*(x)^2) + exp(-80*(x-1)^2)
dens_2 = @. exp(-80*(x-0.5)^2)

fig = figure(figsize=(3, 2))
plot(x, dens_1)
plot(x, dens_2, "--")
xlabel(L"x / \lambda")
ylabel(L"|\psi(x)|^2 / 1/\lambda")
yticks([], []);
```
