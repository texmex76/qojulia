```julia
using PyPlot, Printf
```

```julia
rc("text", usetex=true)
rc("text.latex", preamble= "\\usepackage{amsmath}")
rc("font", family="serif")
pygui(false)

k = 2 * π
arr_length = 100
x = collect(range(0, 1, length=arr_length))
dens_1 = @. sin(k*x)^2
len_new = trunc(Int, arr_length / 2)
dens_1 = vcat(vcat(dens_1[1:len_new-1], zeros(len_new)), dens_1[len_new:end])
# dens_1 = vcat(dens_1[1:50], [1, 2, 3, 4])
# thingy = vcat([1, 2, 3, 4], [1, 2, 3])
arr_length = trunc(Int, 1.5*arr_length)
x_new = collect(range(0, 1, length=arr_length))

plot(x_new, dens_1);
println(size(dens_1))
println(size(x_new))
```
